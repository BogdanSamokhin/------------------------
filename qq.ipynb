```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N = 50
beta = 1
alpha_kvadrat = 0.5
x0 = 0
y0 = 0
y1 = 0
a = np.full(N,np.NAN)
a_point = np.full(N,np.NAN)
a_2point = np.full(N,np.NAN)
b = np.full(N,np.NAN)
c = np.full(N,np.NAN)
gamma = np.full(N,np.NAN)
```

```python
for i in range(25):
    gamma[2 * i] = ((-1) ** i) * (np.sin(x0)/np.math.factorial(2 * i))
    gamma[2 * i + 1] = ((-1) ** i) * (np.cos(x0)/np.math.factorial(2 * i + 1))
```

```python
a[0] = 0
a[1] = 1
b[0] = np.sin(a[0])
c[0] = np.cos(a[0])
b[1] = 1
c[1] = 0
for i in range(2,N):
    a_2point[i] = beta * gamma[i - 2] - alpha_kvadrat * b[i - 2]
    a_point[i] = a_2point[i] / (i - 1)
    a[i] = a_point[i] / i
    b[i] = (a_point[i] * c[i - 1]) / i
    c[i] = ((-1) * a_point[i] * b[i - 1]) / i
```

```python
arr = list(zip(a, a_point,a_2point,b,c,gamma))
df = pd.DataFrame(arr, columns=['a', 'a_point', 'a_2point', 'b', 'c', 'gamma'])
```

```python

x = np.linspace(0, 10, 1000)

# Задаем количество членов суммы степенного ряда, которые будем использовать для построения графика
n_terms = 5

# Вычисляем сумму степенного ряда для каждого значения x
y0 = np.zeros_like(x)
for i in range(1000):
    for n in range(N):
        y0[i] += a[n] * (x[i] ** n)

# Строим график
plt.plot(x, y0)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y')
plt.show()
```
